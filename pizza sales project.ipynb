```python
def category_price(PRICE):
    if PRICE < 10:
        return 'low'
    elif PRICE > 10 and PRICE < 15:
        return 'Medium'
    else:
        return 'High'
```

```python
df['Price Category'] = df['PRICE'].apply(category_price)
```

```python
def market(QUANTITY):
    if QUANTITY<5:
        return 'nice quality'
    elif QUANTITY<8:
        return 'good'
df['market']=df['QUANTITY'].apply(market)
df
```
